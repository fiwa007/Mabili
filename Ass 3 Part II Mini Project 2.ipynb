```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv('telco.csv')

# Display the first few rows of the DataFrame to verify that it has been loaded correctly
df.head()
```

```python
# Summary statistics
print(df.describe())

# Overview of the dataset
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))

# Plot histograms for numerical variables
sns.histplot(df['Age'], ax=axes[0, 0], kde=True)
axes[0, 0].set_title('Distribution of Age')

sns.histplot(df['Tenure in Months'], ax=axes[0, 1], kde=True)
axes[0, 1].set_title('Distribution of Tenure in Months')

sns.histplot(df['Monthly Charge'], ax=axes[0, 2], kde=True)
axes[0, 2].set_title('Distribution of Monthly Charge')

sns.histplot(df['Total Charges'], ax=axes[1, 0], kde=True)
axes[1, 0].set_title('Distribution of Total Charges')

sns.histplot(df['Total Long Distance Charges'], ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Distribution of Total Long Distance Charges')

sns.histplot(df['Total Revenue'], ax=axes[1, 2], kde=True)
axes[1, 2].set_title('Distribution of Total Revenue')

sns.histplot(df['Satisfaction Score'], ax=axes[2, 0], kde=True)
axes[2, 0].set_title('Distribution of Satisfaction Score')

sns.histplot(df['Churn Score'], ax=axes[2, 1], kde=True)
axes[2, 1].set_title('Distribution of Churn Score')

sns.histplot(df['CLTV'], ax=axes[2, 2], kde=True)
axes[2, 2].set_title('Distribution of CLTV')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))

# Plot bar charts for categorical variables
sns.countplot(x='Gender', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Gender')

sns.countplot(x='Married', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Married')

sns.countplot(x='Internet Service', data=df, ax=axes[0, 2])
axes[0, 2].set_title('Internet Service')

sns.countplot(x='Senior Citizen', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Senior Citizen')

sns.countplot(x='Dependents', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Dependents')

sns.countplot(x='Phone Service', data=df, ax=axes[1, 2])
axes[1, 2].set_title('Phone Service')

sns.countplot(x='Contract', data=df, ax=axes[2, 0])
axes[2, 0].set_title('Contract')

sns.countplot(x='Paperless Billing', data=df, ax=axes[2, 1])
axes[2, 1].set_title('Paperless Billing')

sns.countplot(x='Customer Status', data=df, ax=axes[2, 2])
axes[2, 2].set_title('Customer Status')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Exclude non-numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Plot the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title('Correlation Matrix')

# Show plot
plt.show()
```
